```python
string_columns=["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
number = LabelEncoder()
for i in string_columns:
        salary_train[i]= number.fit_transform(salary_train[i])
        salary_test[i]=number.fit_transform(salary_test[i])
```

```python
x_train=salary_train[colnames[0:13]].values
y_train=salary_train[colnames[13]].values
x_test=salary_test[colnames[0:13]].values
y_test=salary_test[colnames[13]].values
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train = norm_func(x_train)
x_test =  norm_func(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB as MB

M_model=MB()
train_pred_multi=M_model.fit(x_train,y_train).predict(x_train)
test_pred_multi=M_model.fit(x_train,y_train).predict(x_test)
```

```python
train_acc_multi=np.mean(train_pred_multi==y_train)
print("Train Accuracy is:",np.round(train_acc_multi*100,2),"%")
```

```python
test_acc_multi=np.mean(test_pred_multi==y_test)
print("Test Accuracy is:",np.round(test_acc_multi*100,2),"%")
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix1=confusion_matrix(y_test, test_pred_multi)
confusion_matrix1
```

```python
from sklearn.metrics import accuracy_score
print("MultinomialNB Model Accuracy:",np.round(accuracy_score(y_test,test_pred_multi)*100,2),"%")
```

```python
from sklearn.naive_bayes import GaussianNB as GB
G_model=GB()
train_pred_gau=G_model.fit(x_train,y_train).predict(x_train)
test_pred_gau=G_model.fit(x_train,y_train).predict(x_test)
```

```python
train_acc_gau=np.mean(train_pred_gau==y_train)
print("Train Accuracy:",np.round(train_acc_gau*100,2),"%")
```

```python
test_acc_gau=np.mean(test_pred_gau==y_test)
print("Test Accuracy:",np.round(test_acc_gau*100,2),"%")
```

```python
confusion_matrix2=confusion_matrix(y_test, test_pred_gau)
confusion_matrix2
```

```python
print("GaussianNB Model Accuracy:",np.round(accuracy_score(y_test,test_pred_gau)*100,2),"%")
```
