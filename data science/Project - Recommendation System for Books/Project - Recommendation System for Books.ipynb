# **Feature Engineering**

```python
final_data = ['User-ID','Age','Location','Book-Title','Book-Rating','Image-URL-L']
final_dataset = books[final_data]
final_dataset.head()
```

```python
final_dataset['City'] = final_dataset['Location'].str.split(',').str[0]
final_dataset['State'] = final_dataset['Location'].str.split(',').str[1]
final_dataset
```

```python
final_dataset = final_dataset.reindex(columns=['User-ID','Age','City','State','Book-Title','Book-Rating','Image-URL-L'])
final_dataset
```

# **Recommendation System**

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score
from sklearn.metrics.pairwise import cosine_similarity
```

# Popularity Based

```python
average_ratings = final_dataset.groupby('Book-Title')['Book-Rating'].mean()
popular_books = average_ratings.sort_values(ascending=False)

# Select the top n books
n = 10
recommended_books = popular_books.index[:n]
print("Recommendation by Popularity: ", recommended_books)
```

# Collaborative Filtering

```python
x = final_dataset.groupby('User-ID').count()['Book-Rating'] > 150
rated_users = x[x].index
```

```python
filtered_rating = final_dataset[final_dataset['User-ID'].isin(rated_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
ratings_pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
ratings_pt
```

```python
ratings_pt.fillna(0,inplace=True)
ratings_pt
```

```python
similarity_scores = cosine_similarity(ratings_pt)
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(ratings_pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = final_dataset[final_dataset['Book-Title'] == ratings_pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))

        data.append(item)

    return data
```

```python
recommend('From the Corner of His Eye')
```

```python
recommend('Zen and the Art of Motorcycle Maintenance: An Inquiry into Values')
```

```python
recommend('Year of Wonders')
```

# **Model Evaluation**

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import precision_score, accuracy_score, confusion_matrix, classification_report
from math import sqrt
```

# Popularity Based

```python
# Regression evaluation metrics
predicted_ratings_popularity = average_ratings[final_dataset['Book-Title']]

mse_popularity = mean_squared_error(final_dataset['Book-Rating'], predicted_ratings_popularity)
rmse_popularity = np.sqrt(mse_popularity)
r2_popularity = r2_score(final_dataset['Book-Rating'], predicted_ratings_popularity)

print("Mean squared error:", mse_popularity)
print("Root mean squared error:", rmse_popularity)
print("R-squared score:", r2_popularity)
```

```python
# Classification evaluation metrics
y_pred_popularity = predicted_ratings_popularity >= 3.5
y_true_popularity = final_dataset['Book-Rating'] >= 3.5

accuracy_popularity = accuracy_score(y_true_popularity, y_pred_popularity)
print("Accuracy:", accuracy_popularity)

precision_popularity = precision_score(y_true_popularity, y_pred_popularity)
print("Precision:", precision_popularity)

cm_popularity = confusion_matrix(y_true_popularity, y_pred_popularity)
print("Confusion Matrix:\n", cm_popularity)

cr_popularity = classification_report(y_true_popularity, y_pred_popularity)
print("Classification Report:\n", cr_popularity)
```

# Collaborative Filtering

```python
# Regression evaluation metrics
predicted_ratings_collab = final_ratings.groupby('User-ID')['Book-Rating'].transform(np.mean)

mse_collab = mean_squared_error(final_ratings['Book-Rating'], predicted_ratings_collab)
rmse_collab = np.sqrt(mse_collab)
r2_collab = r2_score(final_ratings['Book-Rating'], predicted_ratings_collab)

print("Mean squared error:", mse_collab)
print("Root mean squared error:", rmse_collab)
print("R-squared score:", r2_collab)
```

```python
# Classification evaluation metrics
y_pred_collab = predicted_ratings_collab >= 3.5
y_true_collab = final_ratings['Book-Rating'] >= 3.5

accuracy_collab = accuracy_score(y_true_collab, y_pred_collab)
print("Accuracy:", accuracy_collab)

precision_collab = precision_score(y_true_collab, y_pred_collab)
print("Precision:", precision_collab)

cm_collab = confusion_matrix(y_true_collab, y_pred_collab)
print("Confusion Matrix:\n", cm_collab)

cr_collab = classification_report(y_true_collab, y_pred_collab)
print("Classification Report:\n", cr_collab)
```

```python
data={"Model":pd.Series(["Popularity Based","Collaborative Filtering"]),"Accuracy":pd.Series([accuracy_popularity,accuracy_collab]),"RMSE score":pd.Series([rmse_popularity,rmse_collab]),
      "Precision":pd.Series([precision_popularity,precision_collab])}
value_table=pd.DataFrame(data)
value_table.sort_values("Accuracy")
```

***Inference:***
*Post implementation of above recommendation models, we can conclude that Collaborative Filtering Recommendation System is more preferrable to Popularity Based Recommendation system as it has a higher score in two factors - Accuracy (77.7%) and RMSE score (3.26) though Popularity Based Recommendation system has higher precision (0.65)*
